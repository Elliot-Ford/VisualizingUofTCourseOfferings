```python
import pandas as pd
import re
import json
```

```python
data = pd.read_json("data/2018Courses.json")
data
```

```python
cols = data.columns.tolist()
cols = cols[1:4] + cols[5:6] + cols[8:9] + cols[11:13]
print(cols)
selected_data = data[cols]
selected_data
```

```python
def find_prereq(prereq):
    m = re.search(r'([A-Z]{3}[A-Z0-9][0-9]{2}[A-Z][0-9])', prereq)
    if m:
        return [m.group(0)] + find_prereq(prereq[m.end(0)+1:])
    else:
        return []

def get_nonoffered_courses(courses):
    ret = []
    for course in courses:
        if not (selected_data['code'] == course).any():
            ret.append(course)
    return ret
course_json = {}
course_json['nodes'] = []
course_json['links'] = []
for i, r in selected_data.iterrows():
    course_json['nodes'].append({'id':r['code'], 'campus':r['campus'], 'term':r['term'], 'level':r['level']})
    prereq = find_prereq(r["prerequisites"])
    for j in get_nonoffered_courses(prereq):
        course_json['nodes'].append({'id':j, 'campus':'', 'term':'', 'level':0})
    for j in prereq:
        course_json['links'].append({'source':r['code'],'target':j, 'value':1})
```

```python
with open('data/node_edge.json', 'w') as outfile:
    json.dump(course_json, outfile, indent=2)
```
